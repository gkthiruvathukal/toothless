```python
from PIL import Image, ImageFilter
import numpy as np
import matplotlib
from matplotlib.pyplot import imshow
```

```python
%matplotlib inline
im = Image.open("pythonlogo.jpg")
im2 = im.rotate(20)
imshow(np.asarray(im2))
```

```python
im3 = im2.filter(ImageFilter.GaussianBlur(radius=2))
```

```python
imshow(np.asarray(im3))
```
